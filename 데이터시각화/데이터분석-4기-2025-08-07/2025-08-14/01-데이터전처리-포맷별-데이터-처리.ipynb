```python
%%writefile test.xml
<?xml version="1.0" encoding="utf-8"?>
<world name = "주소록과 네이버 API 서비스키">    
    <people name='김민국' >
        <age>15</age>
        <tall>183cm</tall>
        <add>서울시 서대문구</add>
    </people>    
    <people name='박한솔' >
        <age>30</age>
        <tall>175cm</tall>
        <add>서울시 용산구</add>
    </people>    
    <naver_api>
        <client_id>JXRlq56Db1L0PA</client_id>
        <client_secret>hwiniqqkgvH</client_secret>
    </naver_api>    
</world>
```

```python
import xml.etree.ElementTree as ET

tree = ET.parse("test.xml")
root = tree.getroot()
print(root.tag)
print(root.get('name'))

for child in root:
    print(child.tag, child.attrib)

for age in root.iter("age"):
    print(age.text)

for addr in root.iter("add"):
    print(addr.text)

# world 안에 있는 사람들의 키의 합은?

tall_sum = 0
for tall in root.iter("tall"):
    print(tall.text)
    tall_sum = tall_sum + int(tall.text.replace("cm","")) # "183cm" -> 183
 
print("키의 합:", tall_sum)
    
```

```python
# 웹 크롤링 할때 자주 사용 되는 Python 라이브러리 BeautifulSoup과 Selenium
# BeautifulSoup 라이브러리를 이용하여 xml파일을 아래와 같이 파싱하여 텍스트 데이터를 추출

#pip install beautifulsoup4
from bs4 import BeautifulSoup

# f = open("test.xml","r")
# f.close()

with open("test.xml","r",encoding="utf-8") as f:
    xml_doc = f.read()

soup = BeautifulSoup(xml_doc,"xml")

for people in soup.find_all("people"):
    print(people.get('name'))
    print(people.age.text)
    print(people.tall.text)
    print(people.add.text)
    

# naver_api의 id 와 key 를 출력하시오.
for naver_api in soup.find_all("naver_api"):
    print(naver_api.client_id.text)
    print(naver_api.client_secret.text)   

```

```python
# 눈금선 표시
import numpy as np
x = np.arange(1, 11)

y1 = x ** 2
y2 = x ** 3
y3 = x * 4

plt.plot(x, y1,'ro--', x, y2, 'gs--', x, y3, 'b^--')
plt.legend(['red', 'green', 'blue'], loc='best')

# 눈금선이 표시되도록 설정
# plt.grid(True)

# 눈금선의 스타일 설정(옵션 설정)
# describes a sequence of 5 point and 1 point dashes separated by 2 
# "길이가 5인 선, 그 다음 간격 2, 길이가 1인 선, 그 다음 간격 2"의 패턴이 반복되도록 설정
# which='both' = 주눈금선(major ticks)과 보조눈금선(minor ticks) 모두를 표시
plt.grid(which='major', axis='x', color='green', alpha=0.5, dashes=(5,2,1,2))
plt.grid(which='major', axis='y', color='red', alpha=0.5, linestyle='-.', linewidth=1)

plt.grid(True)
```

```python
# 히스토그램
# 수치형(연속형) 데이터에 적합
# 히스토그램은 연속형 데이터의 분포를 시각화할 때 사용하며, 
# 데이터를 구간(bin)으로 나누고 각 구간에 속하는 데이터의 개수를 막대그래프로 나타냄

# 예시 데이터: 학생들의 시험 점수
scores = [55, 67, 78, 90, 85, 66, 72, 88, 93, 45, 68, 74, 82, 95, 77, 60, 70, 50, 80, 89]

plt.hist(scores,bins=7,color="skyblue",edgecolor='black')


# 제목과 축 레이블
plt.title('시험 점수 분포')
plt.xlabel('점수 구간')
plt.ylabel('학생 수')

# 눈금선 추가
plt.grid(axis='y', linestyle='--', alpha=0.5)
```

```python
# 히스토그램
# 수치형(연속형) 데이터에 적합
# 히스토그램은 연속형 데이터의 분포를 시각화할 때 사용하며, 
# 데이터를 구간(bin)으로 나누고 각 구간에 속하는 데이터의 개수를 막대그래프로 나타냄

# 예시 데이터: 학생들의 시험 점수
scores = [55, 67, 78, 90, 85, 66, 72, 88, 93, 45, 68, 74, 82, 95, 77, 60, 70, 50, 80, 89]

#==============================
counts , bins, patches = plt.hist(scores,bins=10,color="skyblue",edgecolor='black')

# 막대 위에 숫자(학생 수) 표기
for count, patch in zip(counts, patches):
    # patch: 각 막대의 사각형 객체
    x = patch.get_x() + patch.get_width() / 2  # 막대의 중앙
    y = patch.get_height()                     # 막대의 높이
    plt.text(x, y + 0.5, str(int(count)), ha='center', va='bottom', fontsize=10)
    

# x축 눈금을 구간 중간값으로 설정
bin_labels = [f'{int(bins[i])}~{int(bins[i+1])}' for i in range(len(bins)-1)]
bin_centers = 0.5 * (bins[1:] + bins[:-1])  # 각 구간의 중심

# x축 눈금 설정
plt.xticks(bin_centers, bin_labels, rotation=45)
#==============================

# 제목과 축 레이블
plt.title('시험 점수 분포')
plt.xlabel('점수 구간')
plt.ylabel('학생 수')

# 눈금선 추가
plt.grid(axis='y', linestyle='--', alpha=0.5)
```

```python
# 히스토그램
# 수치형(연속형) 데이터에 적합
# 히스토그램으로 얻을 수 있는 통찰

# 데이터가 정규분포를 따르는지 체크 가능
# 편향(skewed) 되어 있는지 (왼쪽/오른쪽) 체크
# 이상치가 존재하는지 체크 유무

np.random.seed(1)
#평균이 100, 표준편차가 15인 정규분포 데이터를 10,000개 생성.
x = np.random.normal(100, 15, 10000)

# 전체 범위에 대해(별도 지정 필요 없음), 50개 계급으로 나눈 히스토그램 그리기
plt.hist(x, bins=50)
plt.grid()
plt.show()
```

```python
# 파이 챠트
# 보통 범주형 데이터의 비율을 보여줄 때 사용
# 전체 = 원의 전체 면적이 100%

# 주의
# 항목수가 4 ~ 6개 정도가 알맞음
# 항목수가 너무 많으면, 비율차이가 작아서 조각들이 전부 비슷해 보임
# 정확한 수치 비교는 막대그래프가 더 나음
 
color = ['yellowgreen', 'gold', 'deepskyblue', 'red', 'lightgrey']
langs = ['Ruby', 'C', 'Java', 'Python', 'HTML5']
students = [7, 20, 15, 40, 10]

#가로 6인치, 세로 4인치 크기의 그래프 영역(Figure)을 생성
plt.figure(figsize=(6,4))

#autopct='%.2f%%': 각 조각에 비율을 소수점 둘째 자리까지 표시
# 40 / 총합 × 100
plt.pie(students,labels=langs,colors=color,autopct='%.2f%%')

plt.title("각 언어별 비율")

#그래프 전체 기준으로 오른쪽 아래 (x=1, y=0) 위치
# (0, 0) = 왼쪽 아래
# (1, 1) = 오른쪽 위
# (1, 0) = 오른쪽 아래

# 현재 전체 figure 객체 plt.gcf()
#transFigure는 좌표계를 **figure 전체(0~1 범위의 상대 위치)**로 지정
plt.legend(loc='lower right',
           bbox_to_anchor=(1, 0), bbox_transform=plt.gcf().transFigure)


```

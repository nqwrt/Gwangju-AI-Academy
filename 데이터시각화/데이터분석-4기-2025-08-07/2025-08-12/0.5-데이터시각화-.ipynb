```python
# # 데이터 시각화
# # 데이터 시각화를 위해선 우선 데이터의 종류를 먼저 이해하는 것이 중요

# # 1. 범주형 데이터


# # 막대 → 선 → 히스토그램 → 박스플롯 → 산점도 → 히트맵 순서로 이해

# 실무에서 가장 많이 쓰는 TOP 10
# 막대그래프 → 비교
# 선그래프 → 추세
# 산점도 → 관계
# 히스토그램 → 분포
# 박스플롯 → 이상치
# 히트맵 → 상관관계
# 누적막대 → 구성비
# 파이차트 → 비율
# 트리맵 → 계층
# 산점도 + 회귀선 → 예측

# 한 줄 규칙

# 비교 → 막대 (범주(카테고리)끼리 크기를 비교)
# 변화 → 선
# 관계 → 산점도
# 분포 → 히스토그램
# 비율 → 파이
# 상관 → 히트맵

# 자주 쓰는 plt
# | 코드              | 의미    |
# | --------------- | ----- |
# | `plt.figure()`  | 그림 생성 |
# | `plt.plot()`    | 선그래프  |
# | `plt.bar()`     | 막대그래프 |
# | `plt.scatter()` | 산점도   |
# | `plt.hist()`    | 히스토그램 |
# | `plt.pie()`     | 파이차트  |
# | `plt.title()`   | 제목    |
# | `plt.xlabel()`  | x축    |
# | `plt.ylabel()`  | y축    |
# | `plt.grid()`    | 격자    |
# | `plt.show()`    | 출력    |


```

```python
# 막대 그래프 
import matplotlib.pyplot as plt

dept = [ "개발","기획", "마케팅","디자인"]

count = [
    50,
    30,
    20,
    15
]

plt.figure(figsize=(8,5))

#막대 챠트
plt.bar(
    dept,
    count
)

plt.title( "부서별 인원")

plt.xlabel(
    "부서"
)

plt.ylabel(
    "인원"
)

plt.show()
```

```python
import matplotlib.pyplot as plt

month = [
    "1월",
    "2월",
    "3월",
    "4월",
    "5월"
]

sales = [
    100,
    130,
    120,
    170,
    210
]

plt.figure(
    figsize=(8,5)
)

plt.plot(
    month,
    sales
)

plt.title(
    "월별 매출 추세"
)

plt.xlabel(
    "월"
)

plt.ylabel(
    "매출"
)

plt.grid()
plt.show()

# 변화의 흐름을 보기 쉽기 때문입니다.
```

```python
# 두 숫자형 변수 사이에 관계가 있는지 확인
import matplotlib.pyplot as plt

study = [
    1,
    2,
    3,
    4,
    5
]

score = [
    55,
    65,
    72,
    80,
    90
]

plt.figure(
    figsize=(7,5)
)

plt.scatter(
    study,
    score
)

plt.title(
    "공부시간과 시험점수"
)

plt.xlabel(
    "공부시간"
)

plt.ylabel(
    "점수"
)

plt.grid()

plt.show()

# 두 숫자형 변수 사이에 관계가 있는지 확인
# ↗ 오른쪽 위 → 양의 관계
# ↘ 오른쪽 아래 → 음의 관계
# ● 랜덤하게 흩어짐 → 관계 약함
```

```python
import matplotlib.pyplot as plt

score = [
60,
65,
68,
70,
72,
75,
78,
80,
80,
82,
84,
85,
87,
90,
95
]

plt.figure(
    figsize=(8,5)
)

plt.hist(
    score,
    bins=4 # 구간
)

plt.title(
    "시험 점수 분포"
)

plt.xlabel(
    "점수"
)

plt.ylabel(
    "학생 수"
)

plt.show()

# 히스토그램(Histogram)은 숫자 데이터가 어디에 많이 몰려 있는지(분포) 를 볼 때 사용합니다.
# 막대 → 그룹 비교
# 히스토그램 → 값 분포 확인
# 학생 점수가 어느 구간에 많이 몰려 있는가?

# 60~69 : 3명
# 70~79 : 4명
# 80~89 : 6명 ← 가장 많음
# 90~99 : 2명
```

```python
# 전체에서 각 항목이 얼마만큼 차지하는지(비율) 를 볼 때 사용

import matplotlib.pyplot as plt

product = [
    "A",
    "B",
    "C"
]

sales = [
    50,
    30,
    20
]

plt.figure(
    figsize=(6,6)
)

plt.pie(
    sales,
    labels=product,
    autopct="%1.1f%%"
)

plt.title(
    "제품별 판매 비율"
)

plt.show()

# 파이차트는 언제 사용할까?

# 좋은 예 ✅

# 전체 인원 중 성별 비율
# 시장 점유율
# 예산 구성
# 제품 판매 비중

# 좋지 않은 예 ❌

# 월별 매출 변화
# 학생 점수 비교
# 카테고리가 너무 많음

# (이런 경우는 선그래프나 막대그래프가 더 좋음)
```

```python
# 여러 변수 사이의 상관관계(Correlation)

# 핵심 질문은:
# 어떤 변수끼리 함께 움직이는가?
# 예제로 공부시간, 수면시간, 시험점수 관계를 보겠습니다.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({
"공부시간":[2,3,4,5],
"수면시간":[8,7,6,5],
"시험점수":[60,70,80,90]
})

corr = df.corr()

plt.figure(
    figsize=(6,5)
)

sns.heatmap(
    corr,
    annot=True
)

plt.title(
    "변수 간 상관관계"
)

plt.show()

# 0.99  → 매우 강한 양의 상관
# 0     → 관계 거의 없음
# -0.99 → 매우 강한 음의 상관

# +1     완전 양의 상관
# +0.7   강한 상관
# +0.3   약한 상관
# 0      관계 없음
# -0.3   약한 음의 상관
# -0.7   강한 음의 상관
# -1     완전 음의 상관

# 좋은 예)
# 나이 ↔ 급여 ↔ 만족도
# 키 ↔ 몸무게
# 광고비 ↔ 매출

# 나쁜 예)
# 월별 매출 변화
# 부서별 인원 비교
```

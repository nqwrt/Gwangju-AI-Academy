##### describe()의 평균(mean), 표준편차(std), 사분위수(Q1, Q2, Q3) 는 단순 숫자가 아니라 데이터의 상태를 해석하는 도구
#### 

````python
import pandas as pd

df = pd.DataFrame({
    "score": [
        55,
        60,
        68,
        72,
        75,
        80,
        82,
        85,
        90,
        95
    ]
})

print(df)
````

````python
# describe() 실행

df.describe().T
````

````python
#### 기술통계

# 시험 응시 인원은 총 10명이다.

# 평균 점수는 76.2점이며,
# 중앙값은 77.5점으로
# 평균과 큰 차이가 없어
# 심한 이상치는 없는 것으로 보인다.

# 최저 점수는 55점,
# 최고 점수는 95점이다.

# 표준편차는 12.29로
# 학생들의 성적 분포가
# 다소 퍼져있는 편이다.

# 상위 25% 학생은
# 84점 이상을 기록하였다.
````

# describe() 통계량 해석 가이드

`describe()`의 **평균(mean), 표준편차(std), 사분위수(Q1, Q2, Q3)** 는 단순 숫자가 아니라 **데이터 상태 해석 도구**임.

같은 시험 점수라도 통계량 조합에 따라 전혀 다른 결론 도출 가능성 존재함.

---

# 케이스 1. 평균 높음 + 표준편차 작음 → 전체적으로 우수

## 점수

```text
80 82 83 84 85 86 87 88 89 90
```

## describe()

```text
mean = 85.4
std = 3.0

Q1 = 83
Q2 = 85.5
Q3 = 88
```

## 해석

* 평균 수준 높음
* 학생 간 점수 차이 작음

## 결론

* 학생 전체 수준 우수 상태
* 특정 학생 집중 현상 아님
* 전반적으로 고른 성과 상태

## 실무 예시

* 서비스 만족도 전반적 우수 상태

---

# 케이스 2. 평균 높음 + 표준편차 큼 → 양극화

## 점수

```text
30 40 50 60 80 90 95 98 100 100
```

## describe()

```text
mean = 74
std = 27

Q1 = 52
Q2 = 85
Q3 = 98
```

## 해석

* 평균 수준 높음
* 개인 간 차이 매우 큰 상태

## 결론

* 상위권과 하위권 분리 현상
* 성과 양극화 상태

## 실무 예시

* 평균 만족도 높지만 일부 고객 불만 존재 상태

---

# 케이스 3. 평균 낮음 + 표준편차 작음 → 전체적으로 어려움

## 점수

```text
40 42 44 45 46 47 48 50 51 52
```

## describe()

```text
mean = 46.5
std = 3.8

Q1 = 44
Q2 = 46.5
Q3 = 49
```

## 해석

* 전체 점수 낮은 상태
* 학생 간 차이 작은 상태

## 결론

* 시험 난이도 높은 상태
* 전체 성과 저조 상태

## 실무 예시

* 서비스 만족도 전반적 저하 상태

---

# 케이스 4. 평균과 중앙값 차이 큼 → 이상치 존재

## 점수

```text
60 62 63 64 65 66 67 68 69 100
```

## describe()

```text
mean = 68.4
Q2 = 65.5
```

## 해석

* 평균이 중앙값보다 높은 상태
* 일부 극단값 존재 가능성

## 결론

* 극단적 고득점 영향 발생
* 이상치 존재 가능성

## 판단 기준

```text
mean ≫ median
→ 오른쪽 꼬리 분포

mean ≪ median
→ 왼쪽 꼬리 분포
```

## 실무 예시

* 소수 VIP 고객 중심 매출 구조

---

# 케이스 5. 사분위수 좁음 → 데이터 안정성

## 점수

```text
70 71 72 73 74 75 76 77 78 79
```

## describe()

```text
Q1 = 72
Q3 = 77

IQR = 5
```

```text
IQR = Q3 − Q1
```

## 해석

* 중간 50% 데이터 집중 상태

## 결론

* 학생 수준 유사 상태
* 변동성 낮은 상태

## 실무 예시

* 제품 품질 균일 상태

---

# 케이스 6. 사분위수 넓음 → 편차 큼

## 점수

```text
20 30 40 50 60 70 80 90 95 100
```

## describe()

```text
Q1 = 42
Q3 = 87

IQR = 45
```

## 해석

* 중간 50% 데이터 분산 상태

## 결론

* 학생 간 격차 큰 상태
* 변동성 높은 상태

## 실무 예시

* 지역별 매출 편차 발생 상태

---

# 케이스 7. 표준편차 큼 + 사분위수 좁음 → 이상치 의심

## 점수

```text
70 71 72 73 74 75 76 77 78 150
```

## describe()

```text
mean = 81.6
std = 큼

Q1 = 72
Q3 = 77
```

## 해석

* 대부분 데이터 집중 상태
* 일부 극단값 영향 발생

## 결론

* 이상치 존재 가능성
* 데이터 입력 오류 가능성

## 실무 예시

* 오입력 데이터 발생 가능성

---

# describe() 실무 해석 순서

```text
① 평균(mean)
↓
② 중앙값(50%)
↓
③ 표준편차(std)
↓
④ 사분위수(Q1~Q3)
↓
⑤ 최소/최대값
↓
⑥ 이상치 여부 판단
↓
⑦ 최종 결론 도출
```

---

# 종합 해석 예시

```text
평균 85점
중앙값 84점
표준편차 작음
Q1~Q3 좁음
```

## 최종 결론

* 학생 전체 수준 우수 상태
* 성과 분포 안정 상태
* 이상치 영향 낮은 상태

# 핵심 정리

통계량은 숫자 자체보다 **데이터의 특징과 상태를 설명하는 해석 도구**임.

실제 분석 목표는 숫자 계산이 아니라 **숫자를 의미 있는 스토리로 변환하는 과정**임.

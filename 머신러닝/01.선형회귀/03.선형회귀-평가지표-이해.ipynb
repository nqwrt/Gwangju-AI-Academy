# 📊 머신러닝 평가 지표 완벽 정리

---

## 🚀 1. 평가 지표란?

```text
모델이 얼마나 잘 예측했는지를 측정하는 기준
```

👉 핵심

```text
문제 유형과 상황에 맞게 지표를 선택해야 한다
```

---

# 🎯 2. 분류 (Classification)

👉 목표: **정답을 맞추는 것**

---

## 🔹 Confusion Matrix

|      | 실제 1 | 실제 0 |
| ---- | ---- | ---- |
| 예측 1 | TP   | FP   |
| 예측 0 | FN   | TN   |

---

## 🔹 Accuracy (정확도)

```text
Accuracy = (TP + TN) / 전체
```

* 전체 중 맞춘 비율
* 가장 직관적

❗ 단점
→ 데이터 불균형에 약함

---

## 🔹 Precision (정밀도)

```text
Precision = TP / (TP + FP)
```

* 맞다고 한 것 중 실제 정답 비율
* FP 줄이는 것이 중요

👉 사용 예: 스팸 필터, 사기 탐지

---

## 🔹 Recall (재현율)

```text
Recall = TP / (TP + FN)
```

* 실제 정답 중 맞춘 비율
* FN 줄이는 것이 중요

👉 사용 예: 암 진단, 이상 탐지

---

## 🔹 F1 Score

```text
F1 = 2 * (Precision * Recall) / (Precision + Recall)
```

* Precision과 Recall의 균형
* 불균형 데이터에서 중요

---

## 🔹 ROC-AUC

* 모델의 분류 성능 평가 (확률 기반)
* 0.5 → 랜덤
* 1.0 → 완벽

---

## 🔹 Precision-Recall Curve

* 불균형 데이터에서 매우 중요
* ROC보다 더 유용한 경우 많음

---

# 🎯 3. 회귀 (Regression)

👉 목표: **얼마나 틀렸는지 측정**

---

## 🔹 MAE

```text
MAE = 평균 |실제 - 예측|
```

* 직관적
* 이상치 영향 적음

---

## 🔹 MSE

```text
MSE = 평균 (실제 - 예측)^2
```

* 큰 오차에 민감

---

## 🔹 RMSE

```text
RMSE = √MSE
```

* 실제 단위로 해석 가능
* 가장 많이 사용

---

## 🔹 R² Score

```text
R² = 1 - (오차 / 전체 분산)
```

* 모델 설명력
* 1 → 완벽
* 0 → 평균 수준
* 음수 → 모델 성능 나쁨

---

# 🎯 4. 군집 (Clustering)

👉 목표: **데이터가 잘 묶였는지 평가**

---

## 🔹 Silhouette Score ⭐

* -1 ~ 1
* 1 → 잘 분리됨
* 0 → 겹침
* 음수 → 잘못 군집

---

## 🔹 Davies-Bouldin Index

* 낮을수록 좋음

---

## 🔹 Calinski-Harabasz Index

* 높을수록 좋음

---

# 🚀 마무리

👉 좋은 모델이란?


# 🚀 📈 선형회귀로 평가 지표 이해

---

## 🎯 1. 문제 상황

👉 공부시간으로 시험 점수를 예측하는 **선형회귀 모델**

| 공부시간 (X) | 실제 점수 (y) | 예측 점수 (ŷ) |
| -------- | --------- | --------- |
| 1        | 50        | 55        |
| 2        | 60        | 65        |
| 3        | 70        | 68        |
| 4        | 80        | 78        |
| 5        | 90        | 88        |

---

## 🎯 2. 오차 (Error)

```text
오차 = 실제값 - 예측값
```

예:

* 50 - 55 = -5
* 60 - 65 = -5
* 70 - 68 = 2

👉 모든 평가 지표의 출발점

---

## 🎯 3. MAE (Mean Absolute Error)

```text
MAE = 평균 |y - ŷ|
```

계산:

* |−5| + |−5| + |2| + |2| + |2| = 16
* MAE = 16 / 5 = **3.2**

✔ 의미
👉 평균적으로 **3.2점 정도 틀림**

---

## 🎯 4. MSE (Mean Squared Error)

```text
MSE = 평균 (y - ŷ)^2
```

계산:

* 25 + 25 + 4 + 4 + 4 = 62
* MSE = 62 / 5 = **12.4**

✔ 특징
👉 큰 오차에 더 큰 패널티

---

## 🎯 5. RMSE (Root MSE)

```text
RMSE = √MSE
```

👉 약 **3.52**

✔ 의미
👉 실제 점수 단위로 해석 가능

---

## 🎯 6. R² Score (결정계수)

```text
R² = 1 - (오차 / 전체 분산)
```

✔ 해석:

* 1 → 완벽
* 0 → 평균 수준
* 음수 → 모델이 평균보다 못함

👉 예: R² = 0.95
→ “데이터의 95%를 설명”

---

## 📊 7. 그래프로 이해

### 📈 회귀선

* 점: 실제 데이터
* 선: 모델 예측

👉 선이 데이터를 잘 따라가면 좋은 모델

---

### 📊 Actual vs Predicted

* X축: 실제값
* Y축: 예측값
* 대각선: 완벽한 예측

👉 점이 선에 가까울수록 성능 좋음

---

### 📉 오차 그래프

* Y값: 오차
* 0에 가까울수록 좋음

👉 퍼질수록 모델 성능 나쁨

---

## 🔥 8. 지표 비교

| 지표   | 특징      | 의미    |
| ---- | ------- | ----- |
| MAE  | 직관적     | 평균 오차 |
| MSE  | 큰 오차 강조 | 패널티   |
| RMSE | 해석 쉬움   | 실무 표준 |
| R²   | 설명력     | 모델 품질 |

---

## 💣 9. 실무 포인트

### ❌ 초보

```text
RMSE 하나만 본다
```

### ✅ 실무

```text
MAE + RMSE + R² 함께 확인
```

---

## 🧠 10. 핵심 개념

```text
MAE = 평균 오차
RMSE = 큰 실수 포함 오차
R² = 설명력
```

---

## 🎯 한줄 정리

```text
회귀 평가 = “얼마나 틀렸고, 얼마나 설명했냐”
```

---

## 🚀 마무리

👉 좋은 모델이란?

```text
오차는 작고, 설명력은 높은 모델
```


````python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# -----------------------------
# 1. 데이터 생성 (공부시간 → 점수)
# -----------------------------
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([50, 60, 70, 80, 90])

# -----------------------------
# 2. 모델 학습
# -----------------------------
model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

# -----------------------------
# 3. 평가 지표 계산
# -----------------------------
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y, y_pred)

print("📊 평가 지표")
print(f"MAE  : {mae:.2f}")
print(f"MSE  : {mse:.2f}")
print(f"RMSE : {rmse:.2f}")
print(f"R²   : {r2:.2f}")

# -----------------------------
# 4. 시각화 1 (회귀선)
# -----------------------------
plt.figure()
plt.scatter(X, y, label="Actual")
plt.plot(X, y_pred, label="Regression Line")
plt.xlabel("Study Time")
plt.ylabel("Score")
plt.title("Linear Regression")
plt.legend()
plt.show()

# -----------------------------
# 5. 시각화 2 (Actual vs Predicted)
# -----------------------------
plt.figure()
plt.scatter(y, y_pred)
plt.plot([y.min(), y.max()], [y.min(), y.max()], linestyle="--")
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted")
plt.show()

# -----------------------------
# 6. 시각화 3 (오차 시각화)
# -----------------------------
errors = y - y_pred

plt.figure()
plt.scatter(X, errors)
plt.axhline(0)
plt.xlabel("Study Time")
plt.ylabel("Error (Actual - Predicted)")
plt.title("Error Distribution")
plt.show()
````

# 🚀 📊 R² (결정계수) 수학적 의미 쉽게 이해하기

---

## 🎯 1. 공식

```
R² = 1 - (모델 오차 / 기준 오차)
```

수식 형태:

```
R² = 1 - ∑(y - ŷ)² / ∑(y - ȳ)²
```

---

## 🎯 2. 구성 요소 이해

### 🔹 모델 오차

```
∑(y - ŷ)²
```

* 모델이 예측한 값과 실제 값의 차이
* 작을수록 좋은 모델

---

### 🔹 기준 오차 (Baseline)

```
∑(y - ȳ)²
```

* 그냥 평균으로 예측했을 때의 오차
* 아무것도 안 한 상태

---

## 🎯 3. 핵심 의미

```
R² = 모델이 평균보다 얼마나 더 잘했는가
```

---

## 🔥 4. 숫자로 이해

### ✅ 좋은 모델

* 모델 오차 = 20
* 기준 오차 = 100

```
R² = 1 - (20 / 100) = 0.8
```

👉 해석
→ 데이터의 80%를 설명

---

### ⚖️ 평균 수준 모델

* 모델 오차 = 100
* 기준 오차 = 100

```
R² = 1 - (100 / 100) = 0
```

👉 해석
→ 평균 예측과 동일

---

### ❌ 나쁜 모델

* 모델 오차 = 150
* 기준 오차 = 100

```
R² = 1 - (150 / 100) = -0.5
```

👉 해석
→ 평균보다 못한 모델

---

## 📊 5. 값의 의미

```
R² = 1   → 완벽한 모델
R² = 0   → 평균 수준
R² < 0   → 평균보다 못함
```

---

## 🧠 6. 직관적 비유

👉 시험 문제로 생각해보기

* 평균 점수로 찍기 = 기준 모델
* 내 모델 = 학생

| R² 값 | 의미           |
| ---- | ------------ |
| 0.8  | 평균보다 훨씬 잘 맞춤 |
| 0    | 평균 찍기와 동일    |
| -1   | 평균보다 더 못 맞춤  |

---

## 📉 7. 핵심 구조

```
분자 = 모델 오차
분모 = 기준(평균) 오차

R² = 1 - (모델오차 / 기준오차)
```

---

## 🎯 8. 실무 해석

* RMSE → 얼마나 틀렸는가
* R² → 얼마나 설명했는가

👉 반드시 함께 사용

---

## 🎯 한줄 정리

```
R² = 모델이 데이터 패턴을 얼마나 설명했는지 나타내는 지표
```

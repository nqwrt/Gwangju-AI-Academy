````python
import seaborn as sns

# 식사금액이 증가하면 팁도 증가하는가?

tips = sns.load_dataset("tips")

sns.scatterplot(
    x="total_bill",
    y="tip",
    data=tips
)
````

````python
sns.regplot(
    x="total_bill",
    y="tip",
    data=tips
)
````

````python
# 실습 5개
# 문제1

# 회귀모델 생성

# 문제2

# 기울기 출력

# 문제3

# 절편 출력

# 문제4

# 회귀식 작성

# 문제5

# 40달러 고객 예측
````

````python
from sklearn.linear_model import LinearRegression

x = tips[["total_bill"]]
y = tips["tip"]

model = LinearRegression()

model.fit(x,y)


pred = model.predict([[40]]) 
print(pred)

# 총 식사금액이 40달러인 고객은
# 약 5.12달러의 팁을 줄 것으로 예측된다.
````

## 최종 결론 예시

```text
total_bill과 tip은 양의 상관관계를 보였다.

선형회귀 분석 결과
식사금액이 증가할수록 팁도 증가하였다.

회귀계수는 약 0.105로 나타났으며,
식사금액이 1달러 증가할 때
팁은 평균적으로 0.105달러 증가하였다.

p-value는 0.05보다 작으므로
귀무가설을 기각한다.

따라서 식사금액은 팁에
통계적으로 유의한 영향을 미친다고 결론 내릴 수 있다.
```

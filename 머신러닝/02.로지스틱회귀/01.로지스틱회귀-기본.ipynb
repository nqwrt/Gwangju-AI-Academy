```python
import pandas as pd

df = pd.DataFrame({
    'study_time': [1, 2, 3, 4, 5, 6],
    'pass':       [0, 0, 0, 1, 1, 1]
})

df
```

```python
from sklearn.linear_model import LogisticRegression

# X, y 분리
X = df[['study_time']]
y = df['pass']

print(X.shape)
print(y.shape)

# 모델 생성
model = LogisticRegression()

# 학습
model.fit(X, y)

# 예측
pred = model.predict(X)

print(pred)

pred = model.predict([[10]]) # 10 시간 일때
print(pred)

pred = model.predict([[10],[12]]) # 10 시간 일때 , 12 시간 일때
print(pred)
```

```python
import matplotlib.pyplot as plt

# 실제 데이터
plt.scatter(df['study_time'], df['pass'], label="Actual")

# 예측 결과
plt.scatter(df['study_time'], pred, label="Predicted", marker='x')

plt.xlabel("study_time")
plt.ylabel("pass")
plt.legend()

plt.title("Logistic Regression (Simple Example)")

plt.show()
```

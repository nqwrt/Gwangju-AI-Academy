# 🚀 📊 혼동 행렬 (Confusion Matrix) 완전 정리

---

## 🎯 1. 정의

```text id="c1"
모델이 실제 정답과 예측 결과를 비교해서
어디서 맞고 어디서 틀렸는지 보여주는 표
```

👉 한국어 이름: **혼동 행렬**

---

# 🍎 2. 예시 상황 (스팸 메일 분류)

* 1 = 스팸
* 0 = 정상

---

## 📊 3. 혼동 행렬 구조

|           | 실제 스팸 (1) | 실제 정상 (0) |
| --------- | --------- | --------- |
| 예측 스팸 (1) | TP 👍     | FP ❌      |
| 예측 정상 (0) | FN ❌      | TN 👍     |

---

# 🎯 4. 각 항목 의미

## ✔ TP (True Positive)

```text id="tp1"
스팸인데 스팸이라고 맞춤
```

---

## ❌ FP (False Positive)

```text id="fp1"
정상인데 스팸이라고 잘못 판단 (오탐)
```

---

## ❌ FN (False Negative)

```text id="fn1"
스팸인데 정상이라고 놓침
```

---

## ✔ TN (True Negative)

```text id="tn1"
정상인데 정상이라고 맞춤
```

---

# 🧠 5. 한눈에 이해

```text id="cm1"
실제:    1   1   0   0
예측:    1   0   1   0
결과:    ✔   ❌  ❌  ✔
```

---

# 💣 6. 핵심 의미

```text id="cm2"
혼동 행렬 = 모델의 "실수 지도"
```

---

# 📊 7. 왜 중요한가?

## ❌ Accuracy만 보면

```text id="cm3"
80% 맞췄다 → 끝
```

---

## ✅ 혼동 행렬 보면

```text id="cm4"
- 스팸 10개 중 3개 놓침
- 정상 5개를 스팸으로 오탐
```

👉 훨씬 정확한 분석 가능

---

# 🔥 8. 여기서 모든 지표가 나온다

| 지표        | 계산식                   |
| --------- | --------------------- |
| Accuracy  | (TP + TN) / 전체        |
| Precision | TP / (TP + FP)        |
| Recall    | TP / (TP + FN)        |
| F1 Score  | Precision + Recall 균형 |

# 🍎 🚀 혼동 행렬 (Confusion Matrix) 간단 예제

---

## 🎯 1. 상황: 시험 합격 예측 AI

👉 AI가 5명의 학생에 대해
“합격(1) / 불합격(0)”을 예측했습니다.

---

## 📊 2. 실제 vs 예측 데이터

| 학생 | 실제      | 예측 | 결과   |
| -- | ------- | -- | ---- |
| A  | 1 (합격)  | 1  | ✔ TP |
| B  | 1 (합격)  | 0  | ❌ FN |
| C  | 0 (불합격) | 1  | ❌ FP |
| D  | 0 (불합격) | 0  | ✔ TN |
| E  | 1 (합격)  | 1  | ✔ TP |

---

## 📊 3. 혼동 행렬

|            | 실제 합격 (1) | 실제 불합격 (0) |
| ---------- | --------- | ---------- |
| 예측 합격 (1)  | TP = 2    | FP = 1     |
| 예측 불합격 (0) | FN = 1    | TN = 1     |

---

## 🧠 4. 의미 해석

### ✔ TP (True Positive)

```text id="tp1"
합격인데 합격이라고 맞춤
```

### ❌ FP (False Positive)

```text id="fp1"
불합격인데 합격이라고 잘못 예측
```

### ❌ FN (False Negative)

```text id="fn1"
합격인데 불합격이라고 놓침
```

### ✔ TN (True Negative)

```text id="tn1"
불합격인데 불합격이라고 맞춤
```

---

## 📊 5. Accuracy 계산

```text id="acc1"
Accuracy = (TP + TN) / 전체
         = (2 + 1) / 5
         = 0.6 (60%)
```

---

## 🎯 6. Precision (정밀도)

```text id="pre1"
Precision = TP / (TP + FP)
          = 2 / 3 ≈ 0.67
```

👉 “합격이라고 예측한 것 중 67%만 맞음”

---

## 🎯 7. Recall (재현율)

```text id="rec1"
Recall = TP / (TP + FN)
       = 2 / 3 ≈ 0.67
```

👉 “진짜 합격 중 67%만 찾아냄”

---

```text id="core1"
- 일부는 잘 맞추지만
- 중요한 일부를 놓치는 모델
```

## 🎯 10. 한줄 핵심

지표 Accuray(정확도)의 함정?

🔥 불균형 데이터
👉 예:

100명 중 95명 = 합격
5명 = 불합격

내가 깨도 짚어서 무조건 100명 다 합격 ===
그래도 정확도는 95 % 가 됨
모델이 아무것도 안 했는데도(내가 깬또만 짚었는데) 95%

한마디로 Accuracy는 데이터가 균형일 때만 의미 있음.

🔥 그래서 필요한 것
Precision → 틀린 양성 줄이기
Recall → 놓친 정답 줄이기
F1 → 균형

🎯 Precision이란?

Precision = "내가 맞다고 한 것의 정확도"

📊 예제

AI가 10개 메일을 “스팸”이라고 했어요

근데 실제는:

진짜 스팸 → 7개 ✅
정상 메일 → 3개 ❌ (잘못 잡음)

📊 계산
Precision = 7 / 10 = 0.7 (70%)


| 지표        | 의미                |
| --------- | ----------------- |
| Accuracy  | 전체 중 얼마나 맞췄냐      |
| Precision | 맞다고 한 것 중 얼마나 맞췄냐 |

Precision이 낮으면 "엉뚱한 걸 많이 맞다고 함"
Precision = "틀린 확신을 줄이는 능력"


📊 Recall이란?

Recall = "진짜 범인 중에서 얼마나 많이 잡았냐"

🍎 상황: 범인 찾기 게임

경찰 AI가 있어요 👮‍♂️
👉 “범인을 찾아라!”

🎯 실제 범인

마을에 범인이 10명 있어요.
실제 범인 = 10명


🤖 AI의 결과

AI가 범인이라고 잡은 사람들:
7명 맞게 잡음 ✅
3명 놓침 ❌ (진짜 범인인데 못 찾음)

📊 Recall이란?
Recall = "진짜 범인 중에서 얼마나 많이 잡았냐"

📊 계산
Recall = 7 / 10 = 0.7 (70%)

| 지표        | 질문                |
| --------- | ----------------- |
| Precision | 내가 맞다고 한 게 진짜 맞냐? |
| Recall    | 진짜 중에 얼마나 찾았냐?    |

🔥 아주 중요한 느낌
Precision

👉 “헛발질 줄이기”

Recall

👉 “놓치는 것 줄이기”


#### F1 Score는 

“Precision + Recall을 같이 보는 점수”

🍎 상황 다시 보기: 범인 찾기 게임

경찰 AI가 있어요 👮‍♂️

Precision = “잡았다고 한 사람 중 진짜 범인 비율”
Recall = “진짜 범인 중 얼마나 많이 잡았냐”

🎯 그래서 등장한 F1 Score
Precision + Recall을 균형 있게 보는 점수

![alt text](image-1.png)

F1 Score = Precision과 Recall의 균형 점수

| 지표        | 의미            |
| --------- | ------------- |
| Precision | 맞다고 한 것의 정확도  |
| Recall    | 실제 중 얼마나 찾았는가 |
| F1        | 둘의 균형         |



````python
from sklearn.metrics import confusion_matrix, accuracy_score

# 실제 값 (정답)
y_true = [1, 1, 0, 0, 1]

# 예측 값 (모델 결과)
y_pred = [1, 0, 1, 0, 1]

# 혼동 행렬
cm = confusion_matrix(y_true, y_pred)

print("Confusion Matrix:")
print(cm)

# 정확도
acc = accuracy_score(y_true, y_pred)
print("Accuracy:", acc)
````

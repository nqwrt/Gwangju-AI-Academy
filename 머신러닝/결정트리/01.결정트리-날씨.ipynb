```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

plt.rc('font', family='NanumGothic')
# 날씨 데이터 (숫자 인코딩)
# 0: 맑음, 1: 흐림, 2: 비
# 0: 낮음/약함, 1: 높음/강함
X = [
    [0, 1, 0],  # 맑음, 높음, 약함
    [0, 0, 0],  # 맑음, 낮음, 약함
    [1, 1, 1],  # 흐림, 높음, 강함
    [2, 1, 0],  # 비, 높음, 약함
]

# 정답 (0: No, 1: Yes)
y = [0, 1, 1, 0]

# 모델 생성
model = DecisionTreeClassifier()

# 학습
model.fit(X, y)

# 예측
print(model.predict([[0, 1, 0]]))
```

```python
# 🌳 트리 시각화
plt.figure(figsize=(10, 6))

plot_tree(
    model,
    #feature_names=["날씨", "습도", "바람"],
    feature_names=["A", "B", "C"],
    class_names=["No", "Yes"],
    filled=True
)

plt.show()
```

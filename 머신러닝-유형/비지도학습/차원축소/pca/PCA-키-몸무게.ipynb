# 🧠 PCA (주성분 분석) - Scikit-learn 예제

## 📌 개요
- PCA는 차원을 줄이는 기법
- 여러 feature를 “정보 손실 최소화”로 압축
- 시각화 및 전처리에 많이 사용

---

## 📦 전체 코드 (주석 포함)

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# 1️⃣ 데이터 생성
X = np.array([
    [170, 65],
    [175, 70],
    [180, 75],
    [185, 80]
])

# 2️⃣ PCA 적용 (1차원으로 축소)
pca = PCA(n_components=1)
X_pca = pca.fit_transform(X)

print("원본 데이터:\n", X)
print("PCA 결과:\n", X_pca)

# 3️⃣ 시각화
plt.figure()
plt.scatter(X[:, 0], X[:, 1])
plt.title("Original Data")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

| 컬럼    | 역할    | 중요도   |
| ----- | ----- | ----- |
| pulse | 운동 강도 | ⭐⭐⭐⭐⭐ |
| time  | 운동 시간 | ⭐⭐⭐⭐  |
| kind  | 운동 종류 | ⭐⭐⭐⭐  |
| diet  | 식단    | ⭐⭐⭐   |
| id    | 식별자   | ❌     |


````python
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# 데이터 로드
df = sns.load_dataset("exercise")

# 전처리
# 전처리
df["time"] = df["time"].map({"1 min": 1, "15 min": 15, "30 min": 30})
df["kind"] = df["kind"].map({
    "rest": 0,
    "walking": 1,
    "running": 2
})

# 🔥 핵심 추가
df["time"] = df["time"].astype(int)
df["kind"] = df["kind"].astype(int)

df = df.dropna()

# feature
X = df[["pulse", "time", "kind"]]

# 스케일링
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 군집화
kmeans = KMeans(n_clusters=3, random_state=42)
df["Cluster"] = kmeans.fit_predict(X_scaled)

# 평균
print(df.groupby("Cluster")[["pulse", "time", "kind"]].mean())

# 분포 (핵심🔥)
print(pd.crosstab(df["Cluster"], df["kind"]))
print(pd.crosstab(df["Cluster"], df["time"]))
````

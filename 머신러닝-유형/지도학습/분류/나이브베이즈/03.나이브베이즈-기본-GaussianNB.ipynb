````python
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# 시험 점수 데이터
X = np.array([
    [90, 85],
    [88, 82],
    [60, 55],
    [65, 58],
    [92, 88],
    [58, 52],
    [85, 80],
    [62, 60]
])

y = np.array([1,1,0,0,1,0,1,0])
````

````python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
````

# 📐 Gaussian Naive Bayes (GaussianNB) 공식 정리

---

# 🧠 1. GaussianNB란?

Gaussian Naive Bayes는  
👉 연속형 데이터(실수값)를 정규분포(가우시안 분포)로 가정하여 확률을 계산하는 분류 모델이다.

---

# 📐 2. Gaussian 확률 밀도 함수 (핵심 공식)

$$
P(x_i \mid y)=\frac{1}{\sqrt{2\pi\sigma_y^2}} \exp\left(-\frac{(x_i-\mu_y)^2}{2\sigma_y^2}\right)
$$

---

# 🧠 3. 전체 Naive Bayes 구조

$$
P(y \mid x) \propto P(y) \prod_{i=1}^{n} P(x_i \mid y)
$$

---

# 📊 4. 기호 설명

| 기호 | 의미 |
|------|------|
| $x_i$ | 입력 feature 값 |
| $y$ | 클래스 (예: 합격/불합격) |
| $\mu_y$ | 해당 클래스 평균 |
| $\sigma_y^2$ | 해당 클래스 분산 |
| $P(x_i|y)$ | 해당 값이 클래스 y일 확률 |

---

# 🧠 5. 핵심 직관

👉 GaussianNB는 이렇게 생각하면 된다:

- 각 값이 **클래스 평균에 얼마나 가까운지** 계산
- 가까우면 확률 ↑
- 멀면 확률 ↓

---

# 🍪 6. 비유

👉 “이 학생 점수는 합격자 평균과 얼마나 비슷한가?”

- 비슷하다 → 합격 가능성 높음
- 다르다 → 합격 가능성 낮음

---

# 🚀 7. 한 줄 요약

👉 Gaussian Naive Bayes는 각 feature가 클래스 평균에서 얼마나 떨어져 있는지를 정규분포로 계산해 확률을 구하는 모델이다.

````python
# 📊 1. 문제 설정 (시험 점수 → 합격 예측)

# 우리가 학습한 데이터:

# ✔ 합격 (y=1)
# [90, 85]
# [88, 82]
# [92, 88]
# [85, 80]

# ✔ 불합격 (y=0)
# [60, 55]
# [65, 58]
# [58, 52]
# [62, 60]

#스텝 2
#합격평균 : 88.75  83.75
​#불합격 평균 : 61.25  56.25

#스테 3
#분산계산
#✔ y=1 (시험1)
# (90-88.75)² + (88-88.75)² + (92-88.75)² + (85-88.75)²
# ≈ 10.25

# 중간 생략

# 🧠 10. 핵심 이해

# GaussianNB는 결국 이거야:

# “평균에 가까운 그룹일수록 확률이 높다”

# 🍪 11. 직관 비유
# 합격 학생 평균: 88점
# 불합격 평균: 60점

# 👉 87점?

# 합격 그룹: 👍 매우 가까움
# 불합격 그룹: 👎 너무 멀음


# 🎯 2. 우리가 예측할 데이터
# x = [87, 83]








model = GaussianNB()
model.fit(X_train, y_train)
````

````python
y_pred = model.predict(X_test)
print("예측 결과:", y_pred)
````

````python
print("Accuracy:", accuracy_score(y_test, y_pred))
````

# 📊 Gaussian Naive Bayes 완전 계산 예제 (Velog용)

> ✔ 숫자 대입으로 이해  
> ✔ 평균 + 분산 + 확률 계산 과정  
> ✔ 왜 이 결과가 나오는지 직관까지 포함  

---

# 📊 1. 문제 설정 (시험 점수 → 합격 예측)

## ✔ 학습 데이터

### ✅ 합격 (y = 1)
```
[90, 85]
[88, 82]
[92, 88]
[85, 80]
```

---

# 🎯 2. 예측할 데이터

x = [87, 83]


---

# 📐 3. Gaussian Naive Bayes 공식

$$
P(x_i \mid y)=\frac{1}{\sqrt{2\pi\sigma_y^2}} \exp\left(-\frac{(x_i-\mu_y)^2}{2\sigma_y^2}\right)
$$

---

## 👉 핵심 의미

- $\mu$ : 평균  
- $\sigma^2$ : 분산  
- 평균과 얼마나 가까운지 확률로 계산  

---

# 🧠 4. Step 1: 클래스별 평균

---

## ✔ y = 1 (합격)

### 시험1 평균
$$
\mu_1 = \frac{90+88+92+85}{4} = 88.75
$$

### 시험2 평균
$$
\mu_2 = \frac{85+82+88+80}{4} = 83.75
$$

---

## ✔ y = 0 (불합격)

### 시험1 평균
$$
\mu_1 = \frac{60+65+58+62}{4} = 61.25
$$

### 시험2 평균
$$
\mu_2 = \frac{55+58+52+60}{4} = 56.25
$$

---

# 📊 5. Step 2: 분산 계산 (간단 버전)

## ✔ y = 1 (시험1)

$$
(90-88.75)^2 + (88-88.75)^2 + (92-88.75)^2 + (85-88.75)^2 \approx 10.25
$$

$$
\sigma^2 \approx 2.56
$$

---

# 🧮 6. Step 3: 확률 계산

---

## ✔ y = 1 계산

### 📌 시험1 (87)

$$
P(87 \mid y=1)=\exp\left(-\frac{(87-88.75)^2}{2 \cdot 2.56}\right)
$$

계산 과정:

- $(87 - 88.75)^2 = 3.06$
- $2\sigma^2 = 5.12$

$$
\exp(-3.06 / 5.12) \approx \exp(-0.60) \approx 0.55
$$

---

### 📌 시험2 (83)

$$
(83 - 83.75)^2 = 0.56
$$

$$
\exp(-0.56 / 5.12) \approx \exp(-0.10) \approx 0.90
$$

---

### ✔ 최종 확률 (y=1)

$$
P(x \mid y=1) = 0.55 \times 0.90 = 0.495
$$

---

# 🔴 7. y = 0 계산

---

## 평균 비교

- $\mu \approx (61, 56)$

---

### 📌 시험1 (87)

$$
(87 - 61.25)^2 \approx 665
$$

$$
\exp(-665 / \text{작은 분산}) \approx 0
$$

---

### 📌 시험2 (83)

$$
(83 - 56.25)^2 \approx 716
$$

$$
\exp(-716 / \text{작은 분산}) \approx 0
$$

---

### ✔ 최종 확률 (y=0)

$$
P(x \mid y=0) \approx 0
$$

---

# 🏁 8. 최종 비교

| 클래스 | 확률 |
|--------|------|
| y=1 | 0.495 |
| y=0 | 0.0000 |

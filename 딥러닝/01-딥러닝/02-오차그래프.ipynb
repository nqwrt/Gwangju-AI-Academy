# 📈 기울기 vs 오차 그래프가 나오는 이유 (왜 포물선인가?)

---

## ✅ 1️⃣ 출발점: 우리가 풀고 싶은 문제

우리는 이런 문제를 풀고 있다:

> “데이터에 가장 잘 맞는 직선을 찾자”

직선의 형태:

```
y = ax + b
```

👉 여기서 핵심은
**기울기 a를 어떻게 정할 것인가?**

---

## ✅ 2️⃣ 기준 필요 → 손실함수 등장

좋은 직선인지 판단하려면 기준이 필요하다.

각 점에서의 오차:

```
오차 = 실제값 - 예측값
```

전체 데이터 기준:

```
SSE = ∑ (yᵢ - (a xᵢ + b))²
```

👉 이걸 **손실함수(Loss Function)**라고 한다.

---

## ✅ 3️⃣ 핵심 사고 전환: a를 변수로 본다

* x, y → 주어진 데이터 (고정값)
* b → 평균으로 정리 가능
* 👉 a만 바뀌는 값

즉,

> SSE는 "a에 대한 함수"가 된다

---

## ✅ 4️⃣ 왜 2차 함수가 되는가?

식의 내부를 보면:

```
(yᵢ - (a xᵢ + b))²
```

전개하면:

```
= (yᵢ - b - a xᵢ)²
```

👉 a가 포함된 항을 제곱하고 있음

그래서 전체 합은:

```
SSE(a) = A a² + B a + C
```

---

## 💡 결론

👉 a²이 포함되므로
**무조건 2차 함수(포물선)**

---

## 📊 그래프 형태

* 기울기 작음 → 오차 큼
* 중간 값 → 오차 최소 ⭐
* 기울기 큼 → 오차 다시 증가

👉 결과:

> 📈 U자 모양 그래프 (포물선)

---

## ✅ 5️⃣ 직관적으로 이해

기울기를 계속 바꿔보면:

* 너무 완만 → 데이터 못 맞춤
* 적절한 기울기 → 가장 잘 맞음
* 너무 가파름 → 다시 오차 증가

👉 그래서 자연스럽게:

> “중간에 최소값이 있는 구조”

---

## ✅ 6️⃣ 왜 중요한가 (진짜 핵심)

이 구조 덕분에:

* 최소값이 하나 존재
* 미분으로 해 찾기 가능
* 경사하강법(Gradient Descent) 사용 가능
* 머신러닝 학습 가능

---

## 🔥 핵심 요약

👉 **오차를 제곱했기 때문에 → 기울기에 대한 함수가 2차식이 되고 → 포물선이 된다**

---

## 🎯 추가 포인트 (중요)

만약 오차를 제곱하지 않았다면:

* 절댓값 사용 → V자 그래프
* 세제곱 사용 → 비대칭 그래프

👉 그래서 **제곱을 사용하는 것이 매우 중요한 선택**

---


````python
import numpy as np
import matplotlib.pyplot as plt

# 데이터
x = np.array([2, 4, 6, 8], dtype=float)
y = np.array([81, 93, 91, 97], dtype=float)

# b는 평균으로 고정 (중요 포인트)
mx = np.mean(x)
my = np.mean(y)

# 다양한 기울기 값 생성
a_list = np.linspace(0, 5, 100)

# SSE 저장
sse_list = []

for a in a_list:
    b = my - a * mx  # b는 항상 평균 기반으로 계산
    y_pred = a * x + b
    sse = np.sum((y - y_pred)**2)
    sse_list.append(sse)

# 그래프
plt.figure()
plt.plot(a_list, sse_list)
plt.xlabel("slope (a)")
plt.ylabel("SSE (error)")
plt.title("Slope vs Error (SSE)")
plt.show()
````

# 📊 최소제곱법 vs 경사하강법 (왜 둘 다 필요할까?)

---

## ✅ 결론 먼저

> **작은 문제는 최소제곱(해석적 해)**
> **큰 문제/복잡한 모델은 경사하강법 사용**

---

## ✅ 1️⃣ 최소제곱법 (Closed-form)의 장점

선형회귀는 수식으로 한 번에 해결 가능:

```
a = ∑(xᵢ - x̄)(yᵢ - ȳ) / ∑(xᵢ - x̄)²
```

또는 행렬 형태:

```
θ = (XᵀX)⁻¹ Xᵀy
```

### ✔ 특징

* 한 번에 정답 계산 (반복 없음)
* 정확한 해 (global optimum)

---

## ❌ 2️⃣ 왜 항상 쓰지 않을까?

### 🚨 문제 1: 계산량 폭발

```
(XᵀX)⁻¹
```

👉 역행렬 계산이 매우 비쌈

* 연산량: O(n³)
* 데이터/feature 많으면 계산 불가능 수준

---

### 🚨 문제 2: 메모리 문제

* 행렬 크기가 너무 커짐
* RAM 부족 발생

---

### 🚨 문제 3: 확장성 부족

최소제곱법은 제한적:

* 선형회귀 ✔
* 간단한 모델 ✔

하지만:

* 로지스틱 회귀 ❌
* 신경망 ❌
* 딥러닝 ❌

👉 **닫힌 해(Closed-form)가 없음**

---

## ✅ 3️⃣ 그래서 등장: 경사하강법

아이디어:

> “조금씩 움직이면서 최소값을 찾자”

---

## 🔍 업데이트 공식

```
a := a - α * (d/da) SSE
```

* α (learning rate): 얼마나 이동할지
* 기울기 방향으로 내려감

---

## ✅ 4️⃣ 경사하강법의 장점

### ✔ 대용량 데이터 처리 가능

* 전체 데이터 안 써도 됨 (SGD)

### ✔ 메모리 효율적

* 큰 행렬 필요 없음

### ✔ 확장성 매우 좋음

* 선형회귀
* 로지스틱 회귀
* 딥러닝 (핵심 알고리즘)

---

## 📊 비교 정리

| 항목     | 최소제곱법     | 경사하강법    |
| ------ | --------- | -------- |
| 계산 방식  | 한 번에 해    | 반복       |
| 속도     | 작은 데이터 빠름 | 큰 데이터 유리 |
| 정확도    | 정확한 해     | 근사       |
| 확장성    | 낮음        | 매우 높음    |
| 딥러닝 사용 | ❌         | ✅        |

---

## 🔥 핵심 한 줄

👉 **최소제곱은 이론적 해법, 경사하강법은 현실적인 해결 방법**

---

## 🎯 직관적 이해

* 최소제곱법 👉 공식으로 바로 계산
* 경사하강법 👉 산을 내려가며 최저점을 찾는 방식

👉 softmax는 뉴런마다 붙는 게 아니라
👉 출력 전체를 하나의 확률 분포로 바꾸는 함수다

![alt text](image-18.png)

| 문제 유형 | 출력층 | activation | loss                 |
| ----- | --- | ---------- | -------------------- |
| 이진 분류 | 1   | sigmoid    | binary_crossentropy  |
| 다중 분류 | N   | softmax    | categorical / sparse |
| 다중 라벨 | N   | sigmoid    | binary_crossentropy  |
| 회귀    | 1   | 없음         | mse / mae            |


categorical / sparse. 차이

👉 차이는 “정답(y)의 형태”다

👉 categorical = 원-핫 인코딩
👉 sparse = 정수 라벨

🔍 1. categorical_crossentropy
✔ 정답 형태 (원-핫 인코딩)
y = [
  [1, 0, 0],  # 클래스 0
  [0, 1, 0],  # 클래스 1
  [0, 0, 1]   # 클래스 2
]

👉 벡터 형태

🔍 2. sparse_categorical_crossentropy
✔ 정답 형태 (정수 라벨)
y = [0, 1, 2]

👉 그냥 숫자

⚠️ 5. 자주 하는 실수

❌ 틀린 조합

y = [0,1,2]
loss = 'categorical_crossentropy'   # ❌ 에러

❌ 반대도 틀림

y = [[1,0,0], ...]
loss = 'sparse_categorical_crossentropy'   # ❌

| 상황            | 추천              |
| ------------- | --------------- |
| 일반적인 경우       | sparse (편함)     |
| 직접 one-hot 처리 | categorical     |
| 클래스 많음        | sparse (메모리 절약) |

🔥 최종 한 줄

👉 둘은 완전히 같은 loss지만, 입력 데이터 형식만 다르다


```python
# ✔ 문제
# 입력: 공부시간, 수면시간
# 출력: 등급 (A, B, C → 3개 클래스)

import numpy as np
import tensorflow as tf

# ---------------------
# 1. 데이터
# ---------------------
x = np.array([
    [2, 1],
    [3, 1],
    [4, 2],
    [6, 2],
    [7, 3],
    [8, 3]
], dtype=float)

# 클래스: 0=C, 1=B, 2=A
y = np.array([0, 0, 1, 1, 2, 2])

# ---------------------
# 2. 모델
# ---------------------

# 소프트맥스 이해 주의
model = tf.keras.Sequential([
    tf.keras.layers.Dense(3, input_dim=2, activation='softmax')
])

# ---------------------
# 3. 컴파일
# ---------------------
model.compile(
    optimizer='sgd', #optimizer='sgd' or 'adam' or 'rmsprop'
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# 옵터마이져 제약이 생기는 경우
# 좋은 질문이야 👍
# 결론부터 깔끔하게 말하면:

# 👉 “절대적인 제약”은 거의 없지만,
# 👉 특정 상황에서는 사실상 ‘선택이 제한되는 경우’는 있다

# 🎯 1. 완전 금지 vs 사실상 제약
# ❌ “이 optimizer 쓰면 안 된다” → 거의 없음
# ✅ “이 상황에서는 이게 훨씬 유리하다” → 많음

# 실전 선택 가이드

# | 상황     | 추천 optimizer   |
# | ------ | -------------- |
# | 대부분 문제 | Adam           |
# | 시계열    | RMSprop / Adam |
# | NLP    | Adam / Adagrad |
# | GAN    | Adam (튜닝 필요)   |
# | 최종 튜닝  | SGD            |


model.summary()

# ---------------------
# 4. 학습
# ---------------------
model.fit(x, y, epochs=1000, verbose=0)

# ---------------------
# 5. 예측
# ---------------------
pred = model.predict(x)
print(pred)

```

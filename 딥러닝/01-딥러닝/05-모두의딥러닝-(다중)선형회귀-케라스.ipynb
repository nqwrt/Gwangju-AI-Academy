## 5-2 텐서플로에서 실행하는 다중 선형회귀 모델

### 1.환경 준비

```python
import numpy as np
import matplotlib.pyplot as plt

#텐서플로의 케라스 API에서 필요한 함수들을 불러 옵니다.
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

### 2. 데이터 준비

```python
x = np.array([[2, 0], [4, 4], [6, 2], [8, 3]])
y = np.array([81, 93, 91, 97]) 
print(x.shape,y.shape)
```

### 3. 모델 실행

```python
model = Sequential()

#입력 변수가 2개(학습 시간, 과외 시간)이므로 input_dim에 2를 입력합니다. 
model.add(Dense(1, input_dim=2, activation='linear'))
model.compile(optimizer='sgd' ,loss='mse')

model.fit(x, y, epochs=2000)
```

### 4. 모델 테스트

```python
# 임의의 학습 시간과 과외 시간을 집어넣어 점수를 예측하는 모델을 테스트해 보겠습니다.

# hour = 7
# private_class = 4
# prediction = model.predict([[hour, private_class]])

# print("%.f시간을 공부하고 %.f시간의 과외를 받을 경우, 예상 점수는 %.02f점입니다" % (hour, private_class, prediction))

import numpy as np

hour = 7
private_class = 4

x_test = np.array([[hour, private_class]], dtype=float)

prediction = model.predict(x_test)
print(prediction)
print(f"{hour}시간 공부 + {private_class}시간 과외 → 예상 점수: {prediction[0][0]:.2f}점")
```

2️⃣ 문제: 광고비 + 할인율 → 매출
🧠 문제
x = [[1, 10],
     [2, 20],
     [3, 30],
     [4, 40]]

y = [100, 180, 260, 340]

👉 [5, 50]일 때 매출 예측

📌 설명
광고비 증가 → 매출 증가
할인율 증가 → 매출 증가 효과
✏️ 정답

👉 약 420

💻 코드
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x = np.array([[1, 10],
              [2, 20],
              [3, 30],
              [4, 40]])

y = np.array([100, 180, 260, 340])

model = Sequential()
model.add(Dense(1, input_shape=(2,)))

model.compile(optimizer='sgd', loss='mse')
model.fit(x, y, epochs=1000, verbose=0)

print(model.predict([[5, 50]]))


5️⃣ 문제: 운동 + 식단 → 체중 감소량
🧠 문제
x = [[1, 2000],
     [2, 1800],
     [3, 1500],
     [4, 1200]]

y = [1, 2, 4, 6]

👉 [5, 1000]일 때 체중 감소량 예측

📌 설명
운동 ↑ → 체중 감소 ↑
식단 칼로리 ↓ → 체중 감소 ↑
✏️ 정답

👉 약 8~9

💻 코드
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x = np.array([[1, 2000],
              [2, 1800],
              [3, 1500],
              [4, 1200]])

y = np.array([1, 2, 4, 6])

model = Sequential()
model.add(Dense(1, input_shape=(2,)))

model.compile(optimizer='sgd', loss='mse')
model.fit(x, y, epochs=1000, verbose=0)

print(model.predict([[5, 1000]]))
